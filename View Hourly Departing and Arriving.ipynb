```python
def plot_dep(day,month,year,STN,data):
    y_m_d = '%d-%.2d-%.2d'%(year,month,day)
    single_day = data.loc[y_m_d]

    #select only those arriving at stations STN (STN is a 4 letter code per station)
    dep_stn = single_day.loc[single_day[2] == STN] #column 2 = departing ; column 3 = arriving

    dep_counts = []

    for t in range(24):
        tme = dep_stn.loc[dep_stn[1] == t]
        val = tme[4].sum()
        dep_counts.append(val)
    
    fig = plt.figure()
    plt.plot(range(24), dep_counts)
    plt.xlabel('Hour')
    plt.ylabel('Counts')
    plt.title('The amount of Passengers Departing %s on %d/%d/%d'%(STN,month,day,year))
    plt.figure()
```

```python
def plot_ariv(day,month,year,STN,data):
    y_m_d = '%d-%.2d-%.2d'%(year,month,day)
    single_day = data.loc[y_m_d]

    #select only those arriving at stations STN (STN is a 4 letter code per station)
    dep_stn = single_day.loc[single_day[3] == STN] #column 2 = departing ; column 3 = arriving

    dep_counts = []

    for t in range(24):
        tme = dep_stn.loc[dep_stn[1] == t]
        val = tme[4].sum()
        dep_counts.append(val)
    
    fig = plt.figure()
    plt.plot(range(24), dep_counts)
    plt.xlabel('Hour')
    plt.ylabel('Counts')
    plt.title('The amount of Passengers Arriving at %s on %d/%d/%d'%(STN,month,day,year))
    plt.figure()
```
